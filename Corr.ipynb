```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data=pd.read_csv('ad_saless.csv')
data.head()
```

### Relationship check

```python
corrmatrix=data.corr()
plt.subplots(figsize=(10,6))
sns.heatmap(corrmatrix,vmin=0.9,vmax=1,annot=True,linewidth=0.2,cmap='YlGnBu')
```

There is a strong linear relationship between the amount spent for advertising through five different advertaisements and profit generated

### Correlation and prediction

```python
y=data['Profit(in Thousands)']
x=data.drop(['Date','Profit(in Thousands)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

lr=linear_model.LinearRegression()
model=lr.fit(X_train,Y_train)
predictions=model.predict(X_test)
```

```python
print(predictions)                           
```

```python
mean_squared_error(predictions,Y_test)      #mean squared error
```
