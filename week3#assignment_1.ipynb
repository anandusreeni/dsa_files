## Reading Iris Data Set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
data=pd.read_excel('iris.xls')
```

## Display columns

```python
data.columns
```

## Mean of Each Columns

```python
data.mean()
```

## Null Values

```python
print("Null values Count")
data.isnull().sum()
```

## Visualization

```python
sns.set_style(style='white')
sns.pairplot(data, hue='Classification', size=2.5);
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=data['SL'], y=data['SW'])
plt.show
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x=data['Classification'], y=data['PW'])
plt.show
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(data['Classification'],data['PL'])
```

```python
sns.catplot(x="Classification", kind="count", data=data)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data[['SL','SW','PL','PW']])
```
